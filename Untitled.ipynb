```python
from datetime import datetime
def timer(start_time=None):
    if not start_time:
        start_time = datetime.now()
        return start_time
    elif start_time:
        thour, temp_sec = divmod((datetime.now() - start_time).total_seconds(), 3600)
        tmin, tsec = divmod(temp_sec, 60)
        print('\n Time taken: %i hours %i minutes and %s seconds.' % (thour, tmin, round(tsec, 2)))
```

```python
params_grid = {
                "n_estimators": list(range(100, 1100, 100)),
                "max_depth": [2, 5, 10],
                "learning_rate": [1e-4, 1e-3, 1e-2, 1e-1, 0.5, 1.0],
                #"min_child_weight" : [1, 3, 5, 7],
                "gamma": stats.reciprocal(0.001, 0.1),
                "subsample": np.arange(0.1, 1.0, 0.1),
                "colsample_bytree": np.arange(0.1, 1.0, 0.1),
                "scale_pos_weight": [5, 10, 20, 50, 100],
                "n_jobs": [-1],
                "use_label_encoder": [False],
                "random_state": [42]
        }

fit_params = {
                "early_stopping_rounds": 5,
                "eval_metric":["auc"],
                "eval_set": [(X_valid, y_valid)],
                "verbose":3
        }
```

```python
xgb_clf = xgb.XGBClassifier()
start_time = timer(None)

rand_grid = RandomizedSearchCV(xgb_clf, 
                               params_grid, 
                               n_iter=10,
                               cv=5,
                               scoring="recall", 
                               random_state=42,
                               verbose=3,
                               n_jobs=-1)



#pipe = imbalanced_make_pipeline(SMOTE(sampling_strategy=0.1, random_state=42), rand_grid)
rand_grid.fit(X_os, y_os)


preds = rand_grid.predict(X_valid)

f1 = f1_score(y_valid, preds)
precision = precision_score(y_valid, preds)
recall = recall_score(y_valid, preds)
print("F1 Score: %.3f" %f1)
print("Precision: %.3f" %precision)
print("Recall: %.3f" %recall)

timer(start_time)
```

```python
prob_train = rand_grid.predict_proba(X_train)
prob_train = prob_train[:, 1]
fpr, tpr, threshold = metrics.roc_curve(y_train, prob_train)
roc_auc = metrics.auc(fpr, tpr)
```

```python
J = tpr - fpr
ix = argmax(J)
best_thresh = threshold[ix]
print('Best Threshold=%f' % (best_thresh))
```

```python
"""The Geometric Mean or G-Mean is a metric for imbalanced classification that, 
if optimized, will seek a balance between the sensitivity and the specificity.

Sensitivity = tpr
Specificity = 1 - fpr

G-Mean = sqrt(Sensitivity * Specificity)

"""
gmeans = np.sqrt(tpr * (1-fpr))
# locate the index of the largest g-mean
index = np.argmax(gmeans)
#print(index)
print('Best Threshold=%f, G-Mean=%.4f' % (threshold[index], gmeans[index]))
# plot the roc curve for the model
plt.plot([0,1], [0,1], linestyle='--', label='Random')
plt.plot(fpr, tpr, marker='.', markersize = 2, label='XGBoost AUC = %0.2f' % roc_auc)
plt.scatter(fpr[index], tpr[index], marker='o', s = 100, color='black', label='Best Threshold')

plt.legend(loc = 'lower right')
plt.title('Receiver Operating Characteristic Curve fot Training Data')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
# show the plot
plt.show()
```

```python
roc_train_df = pd.DataFrame([tpr, fpr, threshold], index = ['True Positive Rate', 'False Positive Rate', 'Threshold']).T
roc_train_df['True Negative Rate'] = 1 - roc_train_df['False Positive Rate']
roc_train_df['False Negative Rate'] = 1 - roc_train_df['True Positive Rate']
roc_train_df['G-Means'] = np.sqrt(roc_train_df['True Negative Rate'] * roc_train_df['True Positive Rate'])
roc_train_df = roc_train_df[['True Positive Rate', 'True Negative Rate', 'False Negative Rate', 'False Positive Rate',
                            'Threshold', 'G-Means']]
opt_th = roc_train_df[(roc_train_df['False Negative Rate']*100 - roc_train_df['False Positive Rate']*100) < 1].head(10)
```
